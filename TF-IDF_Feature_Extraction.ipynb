```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/sample_data/new_data.csv')
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
df2 = df.dropna()
```

```python
df2.rename(columns = {'class': 'final_class'},inplace = True)
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
inde = df2.text
dep = df2.final_class
X_train, X_test, y_train, y_test = train_test_split(inde,dep,test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
tvec2 = TfidfVectorizer()
clf = SVC()


from sklearn.pipeline import Pipeline
```

```python
model2 = Pipeline([('vectorizer',tvec2),('classifier',clf)])

model2.fit(X_train, y_train)


from sklearn.metrics import confusion_matrix

predictions = model2.predict(X_test)

confusion_matrix(predictions, y_test)
plot_confusion_matrix(model2 ,X_test,y_test,cmap='Blues',display_labels=['label1','label2'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy : ", accuracy_score(predictions, y_test))
print("Precision : ", precision_score(predictions, y_test, average = 'weighted'))
print("Recall : ", recall_score(predictions, y_test, average = 'weighted'))
print("F1-Score : ", f1_score(predictions, y_test, average = 'weighted'))
```
